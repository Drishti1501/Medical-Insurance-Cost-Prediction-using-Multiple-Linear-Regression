# AI-ML Assignment 1
### Predicting Medical Insurance Costs Using Multiple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Ignore warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')
```

## Task 1: Data Understanding
1. Loading the Pandas DataFrame.
2. Showing the top 5 rows.
3. Identifying features and target.

```python
# 1. Load the dataset (Make sure 'insurance.csv' is in your working directory)
try:
    insurance_data = pd.read_csv('insurance.csv')
    # 2. Show top 5 records
    display(insurance_data.head())
except FileNotFoundError:
    print("Error: 'insurance.csv' not found. Please download it from Kaggle and place it here.")
```

**3. Feature Identification:**
- **Numerical Columns:** `age`, `bmi`, `children`
- **Categorical Columns:** `sex`, `smoker`, `region`
- **Target Attribute:** `charges`

## Task 2: Data Preprocessing
- Checking missing values.
- Encoding categorical data.
- 80/20 train-test split.

```python
if 'insurance_data' in locals():
    # Missing value check
    print("Null values in dataset:\n", insurance_data.isnull().sum())

    # Convert categoricals to dummy variables (One-Hot Encoding)
    processed_data = pd.get_dummies(insurance_data, columns=['sex', 'smoker', 'region'], drop_first=True)
    display(processed_data.head())

    # Defining Features (X) and Target (y)
    X = processed_data.drop('charges', axis=1)
    y = processed_data['charges']

    # Train-test split (80% train, 20% test)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
    print(f"\nShape of Training Data: {X_train.shape}")
    print(f"Shape of Testing Data: {X_test.shape}")
```

## Task 3: Model Development
- Initialize Multiple Linear Regression
- Fit the model and predict.

```python
if 'X_train' in locals():
    # Initialize the model
    lr_model = LinearRegression()
    
    # Fit on training data
    lr_model.fit(X_train, y_train)
    
    # Generate predictions on the test set
    predictions = lr_model.predict(X_test)
```

## Task 4: Model Evaluation
- Calculate MAE, MSE, and R-squared.
- Scatter plot for True vs Predicted.
- Model observations.

```python
if 'predictions' in locals():
    # Metric calculations
    err_mae = mean_absolute_error(y_test, predictions)
    err_mse = mean_squared_error(y_test, predictions)
    score_r2 = r2_score(y_test, predictions)
    
    print(f"Mean Absolute Error (MAE): {err_mae:.3f}")
    print(f"Mean Squared Error (MSE): {err_mse:.3f}")
    print(f"R-squared (R2) Score: {score_r2:.4f}")
    
    # Scatter Plot
    plt.figure(figsize=(9, 6))
    plt.scatter(y_test, predictions, alpha=0.6, color='teal', marker='o')
    # Plotting the ideal fit line
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='orange', linestyle='--', linewidth=2.5)
    plt.xlabel('True Medical Charges')
    plt.ylabel('Model Predicted Charges')
    plt.title('True vs Predicted Insurance Costs')
    plt.grid(alpha=0.4)
    plt.show()
```

**Observations:**
1. **Goodness of Fit:** An R² score of ~0.76 indicates that around 76% of the variability in the medical charges is successfully captured by our linear model.
2. **Prediction Accuracy:** The scatter plot illustrates that predictions for lower medical costs (below 15,000) are fairly tight along the trend line. However, the model under-predicts a substantial number of instances where the true medical costs are extremely high.
3. **Error Magnitudes:** The relatively high MSE and MAE values confirm that there are significant prediction errors present, mostly stemming from the high-cost outliers that a purely linear model fails to estimate.

## Task 5: Conclusion
**Key Findings:** The implemented Multiple Linear Regression model acts as a reliable foundational tool for predicting baseline health insurance charges. While it fits well for average individuals, yielding an R-squared of roughly 0.76, the model struggles considerably with high-end outliers.

**Factors Influencing Costs:** Through the modeling process, it's evident that features like smoking status, body mass index (BMI), and patient age are the primary drivers that heavily influence the escalation of insurance premiums.

**Limitation of Linear Regression:** A primary limitation of using Linear Regression here is its strict reliance on linear independence and additivity. Real-world medical risks often compound in non-linear ways (for instance, the amplified health risks of being both a smoker and having a high BMI). A standard linear model cannot naturally capture these complex interactive effects.
